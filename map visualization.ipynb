## Visualization of the weather stations that generated the weather data

```python
import pandas as pd
import numpy as np
import folium
```

```python
eu_map = folium.Map(location=[54.5260, 15.2551], zoom_start=4.2)
eu_map
```

```python
df_weather = pd.read_csv('eu_weather.csv')
df_weather.head()
```

```python
location = df_weather.groupby(['long', 'lat'])['elevat'].mean().to_frame().reset_index()
location.head()
```

```python
# instantiate a feature group for the incidents in the dataframe
locations = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(location.lat, location.long):
    locations.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=1, # define how big you want the circle markers to be
            color='blue',
            fill=True,
            fill_color='blue',
            fill_opacity=0.5
        )
    )

# add locations to map
eu_map.add_child(locations)
```

```python
import io
from PIL import Image

img_data = eu_map._to_png(10)
img = Image.open(io.BytesIO(img_data))
img.save('image.png')
```
