## Weather Data Process
### In this notebook, I collected weather data from the weather staions throughout europe, located their places(countries) using Google map api.
### The explaination of variables in the final dataframe and the future analysis is shown as follows,
max_tem/tmax: The maximum temperature of the day/month/quarter  
min_tem/tmin: The minimum temperature of the day/month/quarter  
elevation/elevat: The elevation of the weather station/(mean elevation)country  
humi: relative humidity of the day(mean if the time unit is month/quarter)  
wind: wind level of the day(mean if the time unit is month/quarter)  
precip: precipitation of the day  
solar: the UV intensity  
high_t: wheather the max temperature of the day is over 28 degree, 1 yes, 0 no. the number of days that the max temperature is over 28 degree in the month/quarter/year

```python
def find_loc(a, b):
    result = gmaps.reverse_geocode((a, b))
    loc = result[2]['formatted_address'] if len(result)>2 else result[0]['formatted_address']
    return loc
```

```python
location['loc'] = location.apply(lambda x: find_loc(x.lat, x.long), axis=1)
```

```python
eu_nations = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia',
             'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia', 'Slovenia',
             'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia',
             'Spain', 'Sweden']
```

```python
eu_weather = weather.loc[weather['loc'].str.contains('|'.join(eu_nations))]
```

```python
eu_weather.head()
```

```python
def find_nation(x, y):
    for i in y:
        if i in x:
            return i
```

```python
eu_weather['nation'] = eu_weather['loc'].apply(lambda x: find_nation(x, eu_nations))
eu_weather
```

```python
eu_weather.head()
```

```python
nation_weather_grouped = eu_weather.groupby(['nation', 'date'])[['max_tem', 'min_tem', 'elevat', 'precip', 'wind', 'humi', 'solar']].mean()
```
